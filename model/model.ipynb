```python

import os

import numpy as np
import pandas as pd

import tensorflow as tf

import autokeras as ak
```

```python
clf = ak.TextClassifier(
    overwrite=True, max_trials=2, metrics=['mae', 'accuracy'], 
)
```

```python
train= pd.read_csv('cleaned_reviews.csv', names= ['Review', 'Rating'], header = None)
x_train = np.array(train.Review)
y_train = np.array(train.Rating)
```

```python
history = clf.fit(x_train, y_train, epochs=10)
```
